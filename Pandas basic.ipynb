Apply some filters on my dataset

```python
import pandas as pd
dataset = pd.read_csv('Dataset.csv')


# Filter rows where 'age' is greater than 30
filtered_data_age = dataset[dataset['age'] > 30]
print("\nRows where age is greater than 30:")
print(filtered_data_age.head())

#  Filter rows where female smoke
filtered_data_sex_smoker = dataset[(dataset['sex'] == 'female') & (dataset['smoker'] == 'yes')]
print("\nRows where sex is female and they smoke:")
print(filtered_data_sex_smoker.head())

# Filter rows where 'bmi' is between 25 and 30
filtered_data_bmi = dataset[(dataset['bmi'] >= 25) & (dataset['bmi'] <= 30)]
print("\nRows where BMI is between 25 and 30:")
print(filtered_data_bmi.head())

#  Filter rows where 'region' is 'southeast' and 'charges' are greater than 5000
filtered_data_region_charges = dataset[(dataset['region'] == 'southeast') & (dataset['charges'] > 5000)]
print("\nRows where region is southeast and charges are greater than 5000:")
print(filtered_data_region_charges.head())
```

Sort dataset

```python
import pandas as pd
dataset = pd.read_csv('Dataset.csv')



# Sort the dataset by 'age' in ascending order
sorted_data_age = dataset.sort_values(by='age')
print("\nDataset sorted by age in ascending order:")
print(sorted_data_age.head())

# Sort the dataset by 'charges' in descending order
sorted_data_charges_desc = dataset.sort_values(by='charges', ascending=False)
print("\nDataset sorted by charges in descending order:")
print(sorted_data_charges_desc.head())

# Sort the dataset by 'region' and then by 'charges' within each region in ascending order
sorted_data_region_charges = dataset.sort_values(by=['region', 'charges'])
print("\nDataset sorted by region and then by charges within each region in ascending order:")
print(sorted_data_region_charges.head())

# Example 4: Sort the dataset by 'sex' in ascending order and 'age' in descending order
sorted_data_sex_age = dataset.sort_values(by=['sex', 'age'], ascending=[True, False])
print("\nDataset sorted by sex in ascending order and age in descending order:")
print(sorted_data_sex_age.head())
```
