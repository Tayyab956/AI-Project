In this file i am going to make graph of my dataset values

First upload my dataset using matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming your dataset is in a CSV file named 'dataset.csv'
# Adjust the file path as needed
file_path = 'Dataset.csv'

# Importing the dataset using pandas
df = pd.read_csv(file_path)

# Displaying the first few rows of the dataset to confirm import
print(df.head())


```

Histogram of age distribution for smokers and non-smokers

```python
smokers_df = df[df['smoker'] == 'yes']

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(df[df['smoker'] == 'yes']['age'], bins=20, edgecolor='black', color='red')
plt.title('Age Distribution of Smokers')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(df[df['smoker'] == 'no']['age'], bins=20, edgecolor='black', color='blue')
plt.title('Age Distribution of Non-Smokers')
plt.xlabel('Age')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

Scatter plot of BMI vs Charges

```python

smokers = df[df['smoker'] == 'yes']
non_smokers = df[df['smoker'] == 'no']

# Scatter plot of BMI vs Charges with separate plots for smokers and non-smokers
plt.figure(figsize=(10, 6))

plt.scatter(smokers['bmi'], smokers['charges'], c='red', alpha=0.5, label='Smokers')
plt.scatter(non_smokers['bmi'], non_smokers['charges'], c='blue', alpha=0.5, label='Non-Smokers')

plt.title('BMI vs Charges')
plt.xlabel('BMI')
plt.ylabel('Charges')
plt.legend()
plt.show()

```

Bar chart of average charges by region

```python
avg_charges_by_region = df.groupby('region')['charges'].mean()

plt.figure(figsize=(10, 6))
avg_charges_by_region.plot(kind='bar', color='green', edgecolor='black')
plt.title('Average Charges by Region')
plt.xlabel('Region')
plt.ylabel('Average Charges')
plt.show()


```

 Count the number of individuals in each region

```python

region_counts = df['region'].value_counts()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(region_counts))))
plt.title('Distribution of Individuals by Region')
plt.show()
```
